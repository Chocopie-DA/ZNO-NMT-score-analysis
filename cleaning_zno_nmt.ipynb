# Функции

```python
def get_type(name): #вытягивание или изменения типа населенного пункта
    if name.startswith('с.') or name.startswith('селище') \
        or name.startswith('с-ще') or name.startswith('смт') or name == 'село':
        return 'село'
    elif pd.isna(name) or name == 'інша країна':
        return name
    else:
        return 'місто'
```

```python
df_2016 = zno_2016[['OutID', 'Birth', 'SexTypeName', 'Regname', 'AreaName', 'TerName',
                    'UkrTestStatus', 'UkrTest', 'UkrBall100', 
                    'MathTestStatus', 'MathTest', 'MathBall100', 
                    'HistTestStatus', 'HistTest', 'HistBall100',
                    'EngTestStatus','EngTest', 'EngBall100', 
                    'FrTestStatus', 'FrTest', 'FrBall100', 
                    'DeuTestStatus', 'DeuTest', 'DeuBall100', 
                    'SpTestStatus', 'SpTest', 'SpBall100', 
                    'PhysTestStatus', 'PhysTest', 'PhysBall100', 
                    'ChemTestStatus', 'ChemTest', 'ChemBall100', 
                    'BioTestStatus', 'BioTest', 'BioBall100', 
                    'GeoTestStatus', 'GeoTest', 'GeoBall100']]
```

```python
df_2016.loc[:, 'UkrLitStatus'], df_2016.loc[:, 'UkrLit'], df_2016.loc[:, 'UkrLitBall100'] = np.nan, np.nan, np.nan
```

```python
df_2016.loc[:, 'TerTypeName'] = df_2016.loc[:, 'TerName'].apply(get_type) #заполняю новое поле
```
