# Функции

```python
def get_tertype(name): #вытягивание или изменения типа населенного пункта
    if name.startswith('с.') or name.startswith('селище') \
        or name.startswith('с-ще') or name.startswith('смт') or name == 'село':
        return 'село'
    elif pd.isna(name) or name == 'інша країна':
        return name
    else:
        return 'місто'
```

```python
def test_status_2022(score):
    if score >= 100:
        return 'Зараховано'
    elif score == 0:
        return 'Не зараховано'
    else:
        return "Не з'явився або анульовано"
```

```python
def type_summary(ser):
    def get_type(x):
        if pd.isna(x):
            return 'Nan'
        else:
            return type(x).__name__
    return ser.apply(get_type).value_counts(dropna=False)
```

```python
def test_status(status):
    if status in ('Отримав результат', 'Зараховано'):
        return 'Зараховано'
    elif status in ('Не склав', 'Не подолав поріг', 'Не зараховано'):
        return 'Не зараховано'
    elif status in ("Не з'явився", 'Анульовано',
                    "Не з’явився", "Не з'явився або анульовано", "Не з’явився(лася)"):
        return "Не з'явився або анульовано"
    else:
        return "Не складав"
```

## Проверка коректности баллов. 

```python
#проверяет колонки с баллами.
for col in df.columns:
    if col.endswith('Ball'):
        print(type_summary(df[col])) #обнаруживает что есть значение Str
```

```python
#понимаем что некоторые числовые значения представлены как str, непорядок =(
df[df['MathBall'].apply(lambda x: isinstance(x, str))].loc[:, 'MathBall']
```

```python
#исправляем
for col in df.columns:
    mask = df.loc[:, col].apply(lambda x: isinstance(x, str)) #маска для str объектов
    if col.endswith('Ball'):
        df.loc[mask, col] = \
        df.loc[mask, col].apply(float)
        #заменяю 0 значения(не сдал) на 99(проходной бал -1) чтобы не влиять на среднее среди участников
        df.loc[df[col] == 0.0, col] = 99
```
